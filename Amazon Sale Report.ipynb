**2 Product Analysis**

A. Analyze Product Categories

Group by Product Category:

```python
product_analysis = df.groupby('Category').agg({'Qty': 'sum', 'Amount': 'sum'}).reset_index()
```

```python
#sort by revenue
product_analysis = product_analysis.sort_values(by='Amount', ascending=False)
print(product_analysis)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sorting the data by revenue (Amount) in descending order and selecting the top 10 categories
top_10_categories = product_analysis.sort_values(by='Amount', ascending=False).head(10)

# Calculating the percentage of total revenue for each category
top_10_categories['Revenue Percentage'] = top_10_categories['Amount'] / top_10_categories['Amount'].sum() * 100

# Plotting the 100% stacked bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='Revenue Percentage', y='Category', data=top_10_categories, color='skyblue')

# Adding titles and labels
plt.title('Top 10 Product Categories by Revenue (100% Chart)')
plt.xlabel('Percentage of Total Revenue')
plt.ylabel('Product Category')

# Add data labels (percentage) to each bar
for index, value in enumerate(top_10_categories['Revenue Percentage']):
    plt.text(value + 1, index, f'{value:.2f}%', va='center', ha='left', color='black', fontweight='bold')

# Show the plot
plt.show()


```

INSIGHTS: The 100% stacked bar chart reveals that T-shirts and Shirt contribute the most to the total revenue, with T-shirts representing 47.68% and Shirt 25.24% of total sales. This indicates that these two categories are the primary drivers of our sales performance.However, categories such as Socks (0.26%) and Shoes (0.10%) contribute relatively less, suggesting opportunities for growth.

**4 Customer Segmentation**

```python
# Add a segmentation column based on each order's Amount
df['Segment'] = pd.cut(df['Amount'], bins=[0, 100, 500, 1000, float('inf')],
                         labels=['Low', 'Medium', 'High', 'Very High'])

# Display the first few rows to check the segmentation
df[['Order ID', 'Amount']].head()
```

```python
# Example: Segment by fulfillment method
fulfillment_segmentation = df.groupby('fulfilled-by').agg({'Amount': 'sum'}).reset_index()
fulfillment_segmentation['Segment'] = pd.cut(fulfillment_segmentation['Amount'], bins=[0, 1000, 5000, 10000, float('inf')],
                                             labels=['Low', 'Medium', 'High', 'Very High'])

fulfillment_segmentation
```
